Mean AUC normalization



```python
# Load mean AUC data from JSON files
CP_auc_mean = json.load(open('CP_auc_mean.json'))
AC_auc_mean = json.load(open('AC_auc_mean.json'))
LL_auc_mean = json.load(open('LL_auc_mean.json'))
MC_auc_mean = json.load(open('MC_auc_mean.json'))
CPC_auc_mean = json.load(open('CPC_auc_mean.json'))
IP_auc_mean = json.load(open('IP_auc_mean.json'))
LLC_auc_mean = json.load(open('LLC_auc_mean.json'))
MCC_auc_mean = json.load(open('MCC_auc_mean.json'))
MRE_auc_mean = json.load(open('MRE_auc_mean.json'))
PR_auc_mean = json.load(open('PR_auc_mean.json'))
PRD_auc_mean = json.load(open('PRD_auc_mean.json'))
PE_auc_mean = json.load(open('PE_auc_mean.json'))

# Function to normalize AUC values
# def normalize_auc(auc_mean, best_algorithms):
#     # Find the AUC of the best algorithm
#     best_algorithm, best_auc = best_algorithms[0]
    
#     # Normalize each value by the best AUC
#     normalized_auc = {key: value / best_auc for key, value in auc_mean.items()}
    
#     return normalized_auc
def normalize_mean_auc(auc_mean):
    mean_values = list(auc_mean.values())
    min_mean_val, max_mean_val = min(mean_values), max(mean_values)
    
    # print("min_mean_val:", min_mean_val)
    # print("max_mean_val:", max_mean_val)
    
    # if max_val == min_val:
    #     # Avoid division by zero
    #     print("equal!")
    #     return {key: 1.0 for key in auc_mean}
    
    return {key: (mean_value - min_mean_val) / (max_mean_val - min_mean_val) for key, mean_value in auc_mean.items()}

# Normalize AUC values for each problem
normalized_CP_auc_mean = normalize_mean_auc(CP_auc_mean)
normalized_AC_auc_mean = normalize_mean_auc(AC_auc_mean)
normalized_LL_auc_mean = normalize_mean_auc(LL_auc_mean)
normalized_MC_auc_mean = normalize_mean_auc(MC_auc_mean)
normalized_CPC_auc_mean = normalize_mean_auc(CPC_auc_mean)
normalized_IP_auc_mean = normalize_mean_auc(IP_auc_mean)
normalized_LLC_auc_mean = normalize_mean_auc(LLC_auc_mean)
normalized_MCC_auc_mean = normalize_mean_auc(MCC_auc_mean)
normalized_MRE_auc_mean = normalize_mean_auc(MRE_auc_mean)
normalized_PR_auc_mean = normalize_mean_auc(PR_auc_mean)
normalized_PRD_auc_mean = normalize_mean_auc(PRD_auc_mean)
normalized_PE_auc_mean = normalize_mean_auc(PE_auc_mean)

# Print or use the normalized AUC means as needed
print("Normalized mean AUC for CP:", normalized_CP_auc_mean)
print("Normalized mean AUC for AC:", normalized_AC_auc_mean)
print("Normalized mean AUC for LL:", normalized_LL_auc_mean)
print("Normalized mean AUC for MC:", normalized_MC_auc_mean)
print("Normalized mean AUC for CPC:", normalized_CPC_auc_mean)
print("Normalized mean AUC for IP:", normalized_IP_auc_mean)
print("Normalized mean AUC for LLC:", normalized_LLC_auc_mean)
print("Normalized mean AUC for MCC:", normalized_MCC_auc_mean)
print("Normalized mean AUC for MRE:", normalized_MRE_auc_mean)
print("Normalized mean AUC for PR:", normalized_PR_auc_mean)
print("Normalized mean AUC for PRD:", normalized_PRD_auc_mean)
print("Normalized mean AUC for PE:", normalized_PE_auc_mean)
```

### Calculate average mean AUC and rms std AUC, and order algorithms starting with the best average mean AUC



```python
mean_dicts_discrete = [normalized_CP_auc_mean, normalized_AC_auc_mean, normalized_LL_auc_mean, normalized_MC_auc_mean,]
std_dicts_discrete = [normalized_CP_auc_std, normalized_AC_auc_std, normalized_LL_auc_std, normalized_MC_auc_std,]

from collections import defaultdict
def total_mean_across_problems(mean_dicts, std_dicts):
    total_means = defaultdict(float)
    total_std_squares = defaultdict(float)
    num_dicts = len(mean_dicts)
    
    for mean_dict, std_dict in zip(mean_dicts, std_dicts):
        for alg in mean_dict:
            total_means[alg] += mean_dict[alg]
            total_std_squares[alg] += std_dict[alg] ** 2
    avg_means = {k: v / num_dicts for k, v in total_means.items()}
    rms_stds = {k: math.sqrt(v / num_dicts) for k, v in total_std_squares.items()}
    return avg_means, rms_stds

total_means_discrete, total_stds_discrete = total_mean_across_problems(mean_dicts_discrete, std_dicts_discrete)
best_algorithms_discrete = sorted(total_means_discrete.items(), key=lambda x: x[1], reverse=True)

sorted_algorithms = sorted(total_means_discrete.items(), key=lambda x: x[1], reverse=True)
sorted_std_dicts_discrete = [total_stds_discrete[alg] for alg, _ in sorted_algorithms]

print("Best algorithms (by mean):")
for alg, mean_val in sorted_algorithms:
    print(f"{alg}: mean={mean_val:.4f}, std={total_stds_discrete[alg]:.4f}")
```
