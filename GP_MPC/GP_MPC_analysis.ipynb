## Tests


### Cart pole continuous



```python
import cartpole_continuous as cartpole_env
            
env = cartpole_env.CartPoleContinuousEnv(render_mode="rgb_array").unwrapped
env.reset()

next_state, reward, terminated, truncated, info = env.step(env.action_space.sample())
print("next_state:", next_state, env.state)
```

### MuJoCo reacher


```python
env = gym.make("Reacher-v5")
print(env.observation_space.low[:8])
obs, info = env.reset(seed=0)
goal_state_seed0 = obs[-2:]

obs, info = env.reset(seed=8)
goal_state_seed8 = obs[-2:]

obs, info = env.reset(seed=15)
goal_state_seed15 = obs[-2:]

print("goal_state_seed0 ", goal_state_seed0)
print("goal_state_seed8 ", goal_state_seed8)
print("goal_state_seed15 ", goal_state_seed15)

```

### Panda reach


```python
env = gym.make("PandaReach-v3")
obs, info = env.reset(seed=0)
goal_state_seed0 = obs['desired_goal']

obs, info = env.reset(seed=8)
goal_state_seed8 = obs['desired_goal']

obs, info = env.reset(seed=15)
goal_state_seed15 = obs['desired_goal']

print("goal_state_seed0 ", goal_state_seed0)
print("goal_state_seed8 ", goal_state_seed8)
print("goal_state_seed15 ", goal_state_seed15)

```
